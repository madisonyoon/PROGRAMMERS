```python
# 에피소드가 가장 많은 드라마 정보를 출력하시오.
count=['Number of Episode'].value_counts(ascending=False)
count.head(1)

d=drama['Number of Episode']==['Number of Episode'].max()
drama[d]

#에피소드가 가장 적은 드라마 정보 출력
d=drama['Number of Episode']==['Number of Episode'].min()
drama[d]

#에피소드가 가장 적은 드라마 이름 출력
d=drama['Number of Episode']==['Number of Episode'].min()
drama[d]['Name']

#에피소드가 가장 적은 드라마 이름, 에피소드 갯수 출력
d=drama['Number of Episode']==['Number of Episode'].min()
drama[d][['Name','Number of Episode']]
```

```python
# 넷플릭스와 tvN에서 둘 다 방영되는 드라마 검색
## 넷플릭스 랑 tvn 둘다 방영드라마 검색
target=['Netflix','tvN']
d=drama['Network'].map(lamba x: all string in target)
drama[d]
```

# 데이터 가공 및 그룹핑

```python
# 방영일(Aired Date), 에피소드 길이(Duration),줄거리(Synopsis),출연자(Cast),태그(Tags) 컬럼 삭제
col=['Aired Dat','Duration','Synopsis','Cast','Tags']
newdrama=drama.drop(col.axis=1) #axis=0 행, axis=1 열
newdrama.columns
```

```python
# 영문컬럼명을 아래와 같이 한글컬럼명으로 변경한다.
newdrama.columns=['이름','개봉년도','요일','에피소드수','네트워크','등급','장르','순위','평점']
newdrama.columns
```

```python
# 장르 개수 컬럼 생성 -그냥 뒤에 새로운 컬럼 붙이면 생김
newdrama['장르개수']=newdrama['장르'].str.count(',')+1
newdrama.head()
```

```python
#네트워크 개수 생성
newdrama['네트워크개수']=newdrama['네트워크'].str.count(',')+1
newdrama.head()
```

```python
# 맷플롯 한글 폰트 깨짐 현상에 대한 해결 방법
import matplotlib as mpl
import matplotlib.pyplot as plt

plt.rc('font', family='NanumBarunGothic')
```

```python
# 개봉년도별 드라마수
import numpy as np
year_Count=newdrama.groupby('개봉년도').size()
plt.figure(figsize=(16,10))
plt.plot(year_Count,color='r',marker='o')
plt.bar(year_Count.index, year_Count, color='g')
plt.xticks(np.arange(2003,2022))
plt.xlabel('Year of release')
plt.ylabel('Count')
plt.grid()
plt.show()
```

```python
# 개봉년도별 평점의 평균에 대한 시각화
import matplotlib.pyplot as plt
import numpy as np

data=newdrama.groupby('개봉년도')['평점'].mean()
plt.figure(figsize=(16,10))
plt.plot(data,color='r',marker='o')
plt.xticks(np.arange(2003,2022))
plt.xlabel('개봉년도')
plt.ylabel('평점의 평균')
plt.grid()
plt.show()
```
